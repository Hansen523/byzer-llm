```python
import byzerllm
import json
import base64
llm = byzerllm.ByzerLLM.from_default_model("openai_speech_to_text")
## data:audio/${tpe};base64,

audio_file = "/Users/allwefantasy/videos/output_audio.mp3"

with open(audio_file, "rb") as f:
    audio = f.read()
    audio = "data:audio/mp3;base64,"+base64.b64encode(audio).decode("utf-8")

v = llm.chat_oai(conversations=[{
    "role": "user",
    "content": json.dumps({        
        "audio": audio        
    },ensure_ascii=False)
}])
print(json.loads(v[0].output))
```

```python
import ffmpeg
input_video = "/Users/allwefantasy/videos/output.mp4"
output_audio = "/Users/allwefantasy/videos/output_audio.mp3"

try:
    # 读取输入视频文件
    stream = ffmpeg.input(input_video)
    
    # 提取前 10 秒音频
    audio = stream.audio.filter('atrim', duration=10)
    
    # 输出到文件
    output = ffmpeg.output(audio, output_audio)
    
    # 运行 FFmpeg 命令
    ffmpeg.run(output)
    
    print("音频提取成功!")
except ffmpeg.Error as e:
    print("An error occurred: ", e.stderr.decode())
```
